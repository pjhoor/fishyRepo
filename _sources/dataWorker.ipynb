```python
### Initialization

# Import the pandas library
import pandas as pd

# Read the .csv file into a pandas DataFrame
Emissions = pd.read_csv('datasets/totalEmissionsPerCountry2000-2020.csv')
Energy = pd.read_csv('datasets/globalDataOnSustainableEnergy2000-2020.csv')
```

```python
# pandas.DataFrame.head(n=5) for the Emissions dataset
print(Emissions.head(n=5))
```

```python
# pandas.DataFrame.head(n=5) for the Energy dataset
print(Energy.head(n=5))
```

```python
def print_unique_values(df: pd.DataFrame, column: str):
    result = []
    unique_values = df[column].unique()
    for value in unique_values:
        result.append(value)
    return result
print_unique_values(Emissions, 'Area')
```

```python
def calculate_median(df: pd.DataFrame, column: str):
    median = df[column].median()
    return median

calculate_median(Energy, 'Land Area(Km2)')
```
